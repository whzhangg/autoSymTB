```python
from automaticTB.functions import get_structure_from_cif_file
import os, numpy as np
from automaticTB.functions import get_InteractingAOSubspaces_from_cluster
from automaticTB.tools import read_yaml, write_yaml
from automaticTB.interaction import CombinedAOSubspaceInteraction

PbTe = get_structure_from_cif_file("ICSD_CollCode38295.cif", rcut = 4.7)

orbitals = {"Pb":"6s6p", "Te": "5s5p"}

all_interaction_pairs = []
for i, cluster in enumerate(PbTe.nnclusters):
    print(cluster.sitesymmetrygroup.groupname)
    cluster.set_symmetry(find_additional_symmetry=True)
    print(cluster)
    temporary_file = os.path.join("tmp_yaml", f"pbte_2nn_tmp_{i+1}.yaml")
    if os.path.exists(temporary_file):
        iaosubspace = read_yaml(temporary_file)
    else:
        iaosubspace = get_InteractingAOSubspaces_from_cluster(cluster, orbitals)
        write_yaml(iaosubspace, temporary_file)
    all_interaction_pairs += iaosubspace

combined = CombinedAOSubspaceInteraction(all_interaction_pairs)

free_aos = combined.free_AOpairs
for i, free in enumerate(free_aos):
    print(f"{i+1:>2d} " + str(free))
```
